# Learning curves

```python
# 20 steps, time limit is 20 ms
with open('pickles/learning_curves_train_time20.pickle', 'rb') as f:
    learning_curves_train = pickle.load(f)
    
with open('pickles/learning_curves_test_time20.pickle', 'rb') as f:
    learning_curves_test = pickle.load(f)
```

```python
len(learning_curves_train)
```

```python
plt.figure(figsize=(18,10))
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 18}

matplotlib.rc('font', **font)
n_test_PC = 22
n_PC = 17

window = 10

mean_train = np.mean(tuple([learning_curves_train[k] for k in range(len(learning_curves_train))]), axis=0) 
mean_train = [np.mean(mean_train[i:i+window])/n_PC for i in range(len(mean_train)-window)]

mean_test = np.mean(tuple([learning_curves_test[k] for k in range(len(learning_curves_test))]), axis=0)
mean_test = [np.mean(mean_test[i:i+window])/n_test_PC for i in range(len(mean_test)-window)]
            
std_train = np.std(tuple([[np.mean(learning_curves_train[k][i:i+window])/n_PC for i in range(len(learning_curves_train[k])-window)] for k in range(len(learning_curves_train))]), axis=0)    

std_test = np.std(tuple([[np.mean(learning_curves_test[k][i:i+window])/n_test_PC for i in range(len(learning_curves_train[k])-window)] for k in range(len(learning_curves_train))]), axis=0) 

plt.plot(mean_train, 
         label = 'WebDB 2017 PC (train)', alpha=1, color = 'C0', linestyle = ':',linewidth=3)

plt.plot(mean_test, 
         label = 'WebDB 2018 PC (test)', alpha=1, color = 'C1',linewidth=2)  

plt.fill_between(range(len(mean_train)), mean_train - std_train,
                     mean_train + std_train, alpha=0.15,
                     color="C0")

plt.fill_between(range(len(mean_test)), mean_test - std_test,
                     mean_test + std_test, alpha=0.15,
                     color="C1")

plt.yticks(np.arange(0.05,0.4,0.05), (100*np.arange(0.05,0.4,0.05)).astype(int))

plt.xlabel('# simulated session', fontsize=20)
plt.ylabel('% PC members discovered', fontsize=20)


plt.legend()
plt.grid()
plt.show()
```

```python
with open("csv/WebDB2018PC.csv") as inp:
    test_PC = []
    for line in inp:
        split_ = line.split(',')
        test_PC.append(int(split_[0]))
        
with open("csv/WebDB2017PC.csv") as inp:
    PC = []
    for line in inp:
        split_ = line.split(',')
        PC.append(int(split_[0]))
```

```python
with open('pickles/WebDB2017_num_users.pickle','rb') as f:
    num_users = pickle.load(f)
    
with open('pickles/WebDB2017_num_targets.pickle','rb') as f:
    num_targets = pickle.load(f)
    
with open('pickles/WebDB2017_num_target_groups.pickle','rb') as f:
    num_target_groups = pickle.load(f)
    
with open('pickles/WebDB2018_num_users.pickle','rb') as f:
    num_users_test = pickle.load(f)
    
with open('pickles/WebDB2018_num_targets.pickle','rb') as f:
    num_targets_test = pickle.load(f)
    
with open('pickles/WebDB2018_num_target_groups.pickle','rb') as f:
    num_target_groups_test = pickle.load(f)
```

```python
n_PC = len(PC)
n_test_PC = len(test_PC)
```
